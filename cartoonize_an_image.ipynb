```python
#Importing required libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#Reading image 
img = cv2.imread("/content/abc.jpg")
from skimage import io 
io.imshow(img)
```

```python
#Converting to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
io.imshow(img)
```

```python
#Detecting edges of the input image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 9)
edges = cv2.adaptiveThreshold(gray, 235, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
io.imshow(edges)
```

```python
#Cartoonifying the image
color = cv2.bilateralFilter(img, 5, 235, 235)
cartoon = cv2.bitwise_and(color, color, mask=edges)
```

```python

io.imshow(cartoon)
```
